```python
import nc5ng
from nc5ng.nc5data import Conversion, RegionData, GMTPlotter
```

## Conversion (Output) Data

Conversion data is output data generated by `nadcon5-ng` using `NADCON5.0` transformation data and utilities. 

This sample image comes preinstalled with a conversion set for evaluation. Other pre-compiled releases can be downloaded from the [nadcon5.0 github releases page](https://github.com/nc5ng/nadcon5-ng/releases)




```python
c = Conversion(region='conus', old_datum='ussd', new_datum='nad27', out_fdir='/nc5dat/out.ussd.nad27.conus.900.0/')
```

Conversion data contains two portions, `Conversion.input_data` which is the source data from the conversion, including the raw data provided by NGS and `Conversion.output_data` which is the converted point, vector, and grid data for the transformation.

During Loading conversion file names are constructed via attributes and coding scheme used by upstream `NADCON5.0` source code, in the `Conversion.output_data` object, these are indexed by filename. Additional utilities are available to construct a file name from attributes (lat/lon/eht/hor, type of filtering, etc.)

```python
# Print List of Names
print("Output Files : ",[str(_) for _ in c.output_data.raw_data.keys()])
# Get point data for a specific conversion file
o = c.output_data['cvacdlon']
```

Output Data are of type `nc5ng.nc5data.PointData` and represent a collection of points with meta-data. `PointData` is indixed by NGS Point ID (`pid`)

```python
print("Indices ", [str(_) for _ in o.indices])
p = o['JU3588']
```

A single point in the point data set contains the  transformatio information and meta-data, as well as a GMT output data query for plotting purposes. 

```python
print(p.gmt_data)
p.data
```

## Plotting

Conversion data can be plotted using convenience functions to call the `GMT/Python` library

Plotting Options, e.g. Projection, Region, colors, etc. can be specified via key word arguments

Default options can also be set by the `Conversion` object itself before or after construction

```python
#Plot Coverage Points for CONUS
p1 = GMTPlotter.plot_conversion(c, coverage='cvacdlat', vector=None, border=None, land="chocolate", rivers="2/blue")
p1.figure.show(width=1000)
```
